```python
# Il faut tout d'abord faire une jointure entre la table population et la table sous nutrition, en ciblant l'année 2017
#Filtrer sur 2017 pour la population
pop_2017 = population[population['Année'] == 2017]
#Filtrer sur 2016-2018 pour la sous_nutrition
sous_nut_2017 = sous_nutrition[sous_nutrition['Année'] == '2016-2018']
#Faire la jointure entre les deux tables
jointure= pd.merge(pop_2017, sous_nut_2017, on= 'Zone', how= 'left')
```

```python
#Affichage du dataset
jointure.head()
```

```python
#Calcul et affichage du nombre de personnes en état de sous nutrition
nb_sous_nutritiion= jointure['sous_nutrition'].sum()
print("Le nombre de personnes en état de sous nutrition en 2017 est de :", nb_sous_nutritiion)
```

<div style="border: 1px solid RGB(0,150,250);" >
<h3 style="margin: auto; padding: 20px; color: RGB(0,150,250); ">3.2 - Nombre théorique de personne qui pourrait être nourries</h3>
</div>

```python
#Combien mange en moyenne un être humain ? Source =>  https://www.who.int/fr/news-room/fact-sheets/detail/malnutrition
moyenne_nourriture_par_personne_et_par_jour_kcal = 2500
moyenne_nourriture_par_personne_et_par_an_kcal = moyenne_nourriture_par_personne_et_par_jour_kcal * 365
print("En moyenne, un être humain mange {} kcal de nourriture par an.".format(moyenne_nourriture_par_personne_et_par_an_kcal))
```

```python
#On commence par faire une jointure entre le data frame population et Dispo_alimentaire afin d'ajouter dans ce dernier la population
population_pop= population[population['Année']==2017][['Zone', 'Population']]
jointure_pop_dispo= pd.merge(dispo_alimentaire, population_pop, on= 'Zone', how= 'left')
jointure_pop_dispo.fillna(0, inplace= True)
```

```python
#Affichage du nouveau dataframe
jointure_pop_dispo.head()
```

```python
#Création de la colonne dispo_kcal avec calcul des kcal disponibles mondialement
jointure_pop_dispo['dispo_kcal']= (jointure_pop_dispo['Disponibilité alimentaire (Kcal/personne/jour)']*jointure_pop_dispo['Population']*365)
print(" le total des kcal disponibles mondialement est de ", jointure_pop_dispo['dispo_kcal'].sum())


```

```python
#Calcul du nombre d'humains pouvant être nourris
nbre_humains_nourris= jointure_pop_dispo['dispo_kcal'].sum() / (moyenne_nourriture_par_personne_et_par_an_kcal )
print("Le nombre d'humains pouvant être nourris avec les kcal disponibles est de :", nbre_humains_nourris)
```

```python
#Calcul de la disponibilité totale
dispo_totale= jointure_pop_dispo['Disponibilité intérieure'].sum()
print("La disponibilité totale est de :", dispo_totale)
```

```python
#création d'une boucle for pour afficher les différentes valeurs en fonction des colonnes aliments pour animaux, pertes, nourritures, 
colonnes_aliments= ['Aliments pour animaux', 'Pertes', 'Nourriture']
for colonne in colonnes_aliments:
    total = dispo_alimentaire[colonne].sum()
    if colonne == 'Disponibilité intérieure':
        print(f"Disponibilité intérieure totale : {total:,.0f} tonnes")
    elif colonne == 'Nourriture':
        print(f"Alimentation humaine (Nourriture)      : {total:,.0f} tonnes ({total / dispo_alimentaire['Disponibilité intérieure'].sum() * 100:.1f} %)")
    elif colonne == 'Aliments pour animaux':
        print(f"Aliments pour animaux                  : {total:,.0f} tonnes ({total / dispo_alimentaire['Disponibilité intérieure'].sum() * 100:.1f} %)")
    elif colonne == 'Pertes':
        print(f"Pertes                                 : {total:,.0f} tonnes ({total / dispo_alimentaire['Disponibilité intérieure'].sum() * 100:.1f} %)")
    
```

```python
#Création d'un dataframe avec la zone, l'année et l'aide alimentaire puis groupby sur zone et année
aide_alimentaire_grouped= aide_alimentaire.groupby(['Zone', 'Année'])['Valeur'].sum().reset_index()
#Affichage du dataframe
aide_alimentaire_grouped.head(20)

```

```python
#On filtre sur le dataframe avec notre liste
aide_alimentaire_top_5 = aide_alimentaire_grouped[aide_alimentaire_grouped['Zone'].isin(top_5_pays_aide)].reset_index(drop=True)
aide_alimentaire_top_5.head(15)


```

```python
# Affichage des pays avec l'aide alimentaire par année
plt.figure(figsize=(12, 6))
sns.lineplot(data=aide_alimentaire_top_5, x='Année', y='Valeur', hue='Zone', marker='o')
plt.title("Aide alimentaire reçue par les 5 pays ayant le plus bénéficié de l'aide alimentaire au fil des années")
plt.xlabel("Année")
plt.ylabel("Aide alimentaire (kg)")
plt.legend(title='Pays')
plt.grid()
plt.show() 


```

<h3 style="margin: auto; padding: 20px; color: RGB(0,150,250); ">3.11 - Exemple de la Thaïlande pour le Manioc</h3>
</div>

```python
#création d'un dataframe avec uniquement la Thaïlande 
thailande = dispo_alimentaire[dispo_alimentaire['Zone'] == 'Thaïlande']
```

```python
#Calcul de la sous nutrition en Thaïlande
# (Utilise l'intervalle 2016-2018 pour approx 2017, comme pour les autres questions)
sous_nut_thai = sous_nutrition[(sous_nutrition['Année'] == '2016-2018') & (sous_nutrition['Zone'] == 'Thaïlande')]
pop_thai = population[(population['Année'] == 2017) & (population['Zone'] == 'Thaïlande')]['Population'].values[0]
nb_sous_nut_thai = sous_nut_thai['sous_nutrition'].values[0]  
print("Le nombre de personnes en état de sous nutrition en Thaïlande en 2017 est de :", nb_sous_nut_thai)
prop_sous_nut_thai = (nb_sous_nut_thai / pop_thai) * 100
print("En Thaïlande, la proportion de personnes en état de sous nutrition en 2017 est de : {:.2f}%".format(prop_sous_nut_thai))
```

```python
# On calcule la proportion exportée en fonction de la proportion
manioc_thai = thailande[thailande['Produit'] == 'Manioc']
prop_export_manioc = (manioc_thai['Exportations - Quantité'].values[0] / manioc_thai['Production'].values[0]) * 100 if manioc_thai['Production'].values[0] > 0 else 0
print("En Thaïlande, la proportion de manioc exportée est de : {:.2f}%".format(prop_export_manioc))
```

```python
#ajouter en dessous toutes les analyses complémtaires


# 1. Disponibilité en protéines : top 10 max/min (similaire à kcal)
dispo_prot_par_pays = dispo_alimentaire.groupby('Zone')['Disponibilité de protéines en quantité (g/personne/jour)'].sum().reset_index()
dispo_prot_par_pays = dispo_prot_par_pays.rename(columns={'Disponibilité de protéines en quantité (g/personne/jour)': 'Dispo_prot_g_pers_jour'})

top_10_prot = dispo_prot_par_pays.sort_values('Dispo_prot_g_pers_jour', ascending=False).head(10)
bottom_10_prot = dispo_prot_par_pays.sort_values('Dispo_prot_g_pers_jour').head(10)

print("Analyse complémentaire 1 : Top 10 pays avec le plus de disponibilité en protéines (g/pers/jour) :")
display(top_10_prot)

print("Bottom 10 pays avec le moins de disponibilité en protéines (g/pers/jour) :")
display(bottom_10_prot)

# 2. Corrélation aide alimentaire totale (2013-2016) et sous-nutrition (2017)
aide_totale = aide_alimentaire.groupby('Zone')['Valeur'].sum().reset_index()
sous_nut_2017 = sous_nutrition[sous_nutrition['Année'] == '2016-2018'][['Zone', 'sous_nutrition']]
corr_aide_sousnut = aide_totale.merge(sous_nut_2017, on='Zone', how='inner').sort_values('Valeur', ascending=False).head(10)

print("\nAnalyse complémentaire 2 : Top 10 pays aidés vs. sous-nutrition (aide en tonnes) :")
display(corr_aide_sousnut)

# 3. Pertes totales mondiales
pertes_totales = dispo_alimentaire['Pertes'].sum()

print("\nAnalyse complémentaire 3 : Pertes alimentaires totales mondiales :")
print(f"{pertes_totales:,.0f} tonnes (soit environ 4.6% de la disponibilité intérieure)")

# 4. Optionnel : Graphique simple (ex. barplot top 10 pertes par pays)
pertes_par_pays = dispo_alimentaire.groupby('Zone')['Pertes'].sum().sort_values(ascending=False).head(10)
plt.figure(figsize=(10,6))
sns.barplot(x=pertes_par_pays.values, y=pertes_par_pays.index, palette='Reds_d')
plt.title('Top 10 pays avec le plus de pertes alimentaires (tonnes)')
plt.xlabel('Pertes (tonnes)')
plt.ylabel('Pays')
plt.grid()
plt.show()
```
